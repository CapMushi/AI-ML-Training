```python
import pandas as pd
import numpy as np

# 1. Load the Spotify structural dataset
df = pd.read_csv('data/spotify.csv')

print("--- Data Dimensionality Check ---")
print(f"Total Song Entries (Rows): {df.shape[0]}")
print(f"Total Structural Features (Columns): {df.shape[1]}\n")

# 2. Separate variables intentionally based on the Data Dictionary definition
metadata_cols = ['track_id', 'track_name', 'track_artist', 'track_album_id', 
                 'track_album_name', 'track_album_release_date', 'playlist_name', 'playlist_id']

categorical_cols = ['playlist_genre', 'playlist_subgenre', 'key', 'mode']

continuous_numeric_cols = ['track_popularity', 'danceability', 'energy', 'loudness', 
                           'speechiness', 'acousticness', 'instrumentalness', 
                           'liveness', 'valence', 'tempo', 'duration_ms']

print("--- Architectural Feature Split ---")
print(f"Metadata String Fields: {len(metadata_cols)}")
print(f"Categorical/Encoded Fields: {len(categorical_cols)}")
print(f"Continuous Numeric Audio Waves: {len(continuous_numeric_cols)}\n")

print("--- Null Value Completeness Audit ---")
null_counts = df.isnull().sum()
null_percentages = (df.isnull().sum() / len(df)) * 100

null_analysis_df = pd.DataFrame({
    'Null Counts': null_counts,
    'Percentage (%)': null_percentages
}).sort_values(by='Null Counts', ascending=False)

if null_counts.sum() == 0:
    print("Dataset completeness: 100%. No missing data discovered.")
else:
    print(null_analysis_df[null_analysis_df['Null Counts'] > 0])
```
